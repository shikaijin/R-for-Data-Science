```r
# Use filter to select flights that departed in Jan or Nov such that the flights were not delayed on arrival
# by more than 10 minutes.
library(nycflights13)
filter(flights, month == 1 | month == 11, arr_delay <= 10)
```

```r
# Write R code to find the number of flights that departed in May.
nrow(filter(flights, month == 5))
```

```r
# Use ggplot to plot a histogram of the departure delay of the flights that departed in Feb.
flights %>% 
  filter(month == 2) %>% 
  ggplot(aes(x = dep_delay)) +
  geom_histogram(binwidth = 10)
```

```r
# Use ggplot to plot a histogram of the departure delay of the flights that departed in Feb with
# departure delay less than 100 minutes.
flights %>% 
  filter(month == 2, dep_delay < 100) %>% 
  ggplot(aes(x = dep_delay)) +
  geom_histogram(binwidth = 5)
```

```r
# a) Use ggplot to create a scatterplot of arr_delay versus dep_delay using the flights that departed
# in Jan 1 with departure delay strictly less than 3 hours.
delay1 <- filter(flights, month == 1, day == 1, dep_delay < 180)
ggplot(delay1, aes(x = dep_delay, y = arr_delay)) +
  geom_point() 
```

```r
# b) Use ggplot to create a scatterplot of arr_delay versus dep_delay using the flights that departed in
# Jan 1 with departure delay strictly less than 10 minutes.
delay2 <- filter(flights, month == 1, day == 1, dep_delay < 10)
ggplot(delay2, aes(x = dep_delay, y = arr_delay)) +
  geom_point() 
```

#### Graph in (a) shows a more apparent trend between the two delay times since graph in (a) shows a stronger positive association than Graph in (b).

```r
# Compute the correlation for the points that you plot in (a) and (b).
# The correlation for the points in (a) is:
cor(delay1$dep_delay, delay1$arr_delay, use="complete.obs")

# The correlation for the points in (b) is:
cor(delay2$dep_delay, delay2$arr_delay, use="complete.obs")
```

```r
# Write a function called plot_delay with inputs which_month, which_day, lower_range and
# upper_range to create the scatterplot (using ggplot) of arr_delay versus dep_delay using the flights
# that departed in month=which_month and day=which_day with departure delay between (inclusively)
# lower_range and upper_range
plot_delay <- function(which_month, which_day, lower_range, upper_range){
  flights %>%
    filter(month == which_month, day == which_day, between(dep_delay, lower_range, upper_range)) %>% 
    ggplot(aes(x = dep_delay, y = arr_delay)) +
    geom_point()
}
plot_delay(which_month = 1, which_day = 30, lower_range = -5, upper_range = 30)
```

```r
# Use ggplot to create a bar chart of counts of the number of flights in each month.
ggplot(flights, aes(x = as.factor(month))) + 
       geom_bar()
```

```r
# Create a line graph of the average departure delay. 
avg_dep_delay <- 
  flights %>% 
  group_by(month, day) %>% 
  summarize(avg_dep_delay = mean(dep_delay, na.rm = TRUE)) %>%
  ungroup() %>% 
  mutate(Time = 1:365)

ggplot(avg_dep_delay, aes(x = Time, y = avg_dep_delay)) +
  geom_line() 
```
